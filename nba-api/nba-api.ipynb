```python
def get_season_stats(season, category):
    season_stats = leaguedashplayerstats.LeagueDashPlayerStats(season=season)
    df = season_stats.get_data_frames()[0]
    df = df[["PLAYER_ID", "PLAYER_NAME", category]]
    return df
```
